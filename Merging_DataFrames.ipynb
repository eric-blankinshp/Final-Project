```python
import numpy as np
import pandas as pd
```

```python
# Importing csv files and assigning their df. 

cleaned_housing_data = pd.read_csv('cleaned_housing.csv')  
cleaned_school_data = pd.read_csv('cleaned_schools.csv')
```

```python
cleaned_housing_data.head()
```

```python
# Dropping columns except for most current housing data
cleaned_housing_data.drop(cleaned_housing_data.iloc[:, 9:280], inplace=True, axis=1)
```

```python
cleaned_housing_data.head()
```

```python
# Oops, forgot one"

cleaned_housing_data = cleaned_housing_data.drop(["8/31/2022"], axis=1)
```

```python
# Renaming Housing Value column
cleaned_housing_data.rename(columns={'9/30/2022': 'Housing Value', 'RegionName': 'ZIP'}, inplace=True)
```

```python
cleaned_housing_data.head()
```

```python
cleaned_school_data.head()
```

```python
# Merging DF's
merged_df = pd.merge(cleaned_housing_data, cleaned_school_data, on='ZIP', how='outer')
```

```python
merged_df.head()
```

```python
for col in merged_df.columns:
    print(col)
```

```python
merged_df = merged_df.drop(["StateName", "X", "Y"], axis=1)
```

```python
# Exporting school csv 

# Create the output file (CSV).
output_data_file = "merged_df"
# Export the City_Data into a CSV.
merged_df.to_csv("merged_df.csv", index=False)
```
