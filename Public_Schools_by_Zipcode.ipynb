```python
import numpy as np
import pandas as pd
```

```python
# Importing csv files and assigning their df. 

school_data = pd.read_csv('Public_Schools_by_Zipcode.csv')  
housing_values = pd.read_csv('Zipcode_HousingValue_Zillow.csv')
```

<h1>School Data DF</h1>

```python
# Checking school DF 
school_data
```

```python
#Checking shape of school df. 
school_data.shape
```

```python
# Checking for null values 

school_data.isnull().sum()
```

```python
# Dropping null values 

school_data.dropna().shape
```

```python
# Checking for Unique values in "State" column. We might have international schools in the dataset. 

print(school_data["STATE"].unique())
```

```python
# Defining list of states 
states = ["PR","GU","HI","VI","AS"]
```

```python
# Removing international schools from dataset. PR (Puerto Rico), GU (Guam), HI (Hawaii), VI (Virgin Islands), AS (American Samoa) 

clean_schools_df = school_data[school_data.STATE.isin(states) == False]
clean_schools_df
```

```python
clean_schools_df.shape
```

```python
# Checking column names to see which we should keep for future use. 
clean_schools_df.columns
```

```python
# Exporting school csv 

# Create the output file (CSV).
output_data_file = "clean_schools_df"
# Export the City_Data into a CSV.
clean_schools_df.to_csv("cleaned_schools.csv", index=False)
```
